```python
# Listing files on a folder

import os

path = '/content/gdrive/MyDrive/TugasAkhirS1-DRL/5TradingViewScraper/5.1.DataPerStock/'
files = os.listdir(path)

for f in files:
	print(f)
```

```python
# import modules
import pandas as pd
import datetime
import time
import numpy as np

file_count = len(files)

# create empty list
dataframes_list = []

# append datasets to the list
for i in range(file_count):
	temp_df = pd.read_csv(path+"/"+files[i])
	dataframes_list.append(temp_df)
```

```python
import time
import datetime
from time import mktime
from datetime import datetime as dt

new_dataframes_list = []

j=0
for df in dataframes_list:
	df.columns = ["tic", "indeks", "datadate", "open", "high", "low", "close", "volume"]
	df.rename(columns = {'close':'adjcp'}, inplace = True)
	df = df[['indeks', 'datadate', 'tic', 'adjcp', 'open', 'high', 'low', "volume"]]
	df['datadate'] = df['datadate'].fillna(0).astype(int)
	arr = df.to_numpy()
	for i in arr:
		arr[j][1] = dt.fromtimestamp(arr[j][1])
		j+=1
	df = pd.DataFrame(arr)
	df.columns = ['indeks', 'datadate', 'tic', 'adjcp', 'open', 'high', 'low', "volume"]
	df['datadate'] = pd.to_datetime(df['datadate']).dt.date
	df['datadate'] = pd.to_datetime(df['datadate'], format='%Y-%m-%d')
	df = df.loc[(df['datadate'] >= '2009-01-01') & (df['datadate'] <= '2020-08-05')]
	df['datadate'] = pd.to_datetime(df['datadate'], errors='coerce')
	df['datadate'] = df['datadate'].dt.strftime('%Y%m%d')
	new_dataframes_list.append(df)
	display(df)
	j=0
```

```python
# Merge the Dataframes
mergedDataframe = pd.concat(new_dataframes_list,ignore_index=True)
display(mergedDataframe)
```

```python
# Save dataframe as csv
dataset = mergedDataframe.to_csv("/content/gdrive/MyDrive/TugasAkhirS1-DRL/5TradingViewScraper/5.2.MergedDataset/TradingView_JIIDataset_2009_2020.csv")
```
